# A line graph to represent the change in the number of rice grains

```python
number_grains = 1
days=100
list_number_grains = [number_grains]
list_total_grains =[number_grains]

for i in range(days-1):
    number_grains *= 2
    list_number_grains.append(number_grains)
    list_total_grains.append(sum(list_number_grains))
    
import matplotlib.pyplot as plt
plt.title("Increase in Number of grains")
plt.xlabel("number of days")
plt.ylabel("nmber of grains")
plt.plot(list_total_grains)
plt.show()
```

# Problem 2

```python
def compute_sorori_shinzaemon(day=150):
    """
    A function that verifies the anecdote of rice by Sorori Shinzaemon. Calculate the number of rice grains per day.

    Parameteres
    --------------
    day: int
        Number of days to get rice (default : 100)

    Returns
    --------------
    list_n_grains : list
        List of the number of rice you get one day (the length of the list is equal to day)
    list_total_grains : list
        List of the total number of rice you will receive by a certain day (the length of the list will be equal to day)
    """
    # Write the code here. pass is an instruction that means do nothing.
    number_grains = 1
    list_n_grains = [number_grains]
    list_total_grains = [number_grains]
    
    for i in range(day-1):
        number_grains *=2
        list_n_grains.append(number_grains)
        list_total_grains.append(sum(list_n_grains))
    
    plt.title("Number of grains pre- Day")
    plt.xlabel("number of day")
    plt.ylabel("number of grains")
    plt.plot(list_n_grains, label='number of grains')
    plt.plot(list_total_grains, label='accumulation of grains',color='green')
    return list_n_grains, list_total_grains
list_n_grains, list_total_grains = compute_sorori_shinzaemon(day=80)
```

```python
def lifespan(grains,people):
   """
   a function that calculate how many people can live for how many days with this rice
   """
    daily_grains_consumption =20000
    days_of_consumption = grains/people*daily_grains_consumption
    return days_of_consumption
```
